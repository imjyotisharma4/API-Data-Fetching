```python
x=requests.get("https://api.themoviedb.org/3/movie/popular?api_key=3b76c943814293d6a2e91698a5a8bfc1&language=en-US&page=1").json()
```

```python
# creating list of the data & converting it to the DataFrame

popularity=[]
vote_count=[]
title=[]
original_language=[]
vote_average=[]
overview=[]
release_date=[]
for i in x['results']:
    popularity.append(i['popularity'])
    vote_count.append(i['vote_count'])
    title.append(i['title'])
    original_language.append(i['original_language'])
    vote_average.append(i['vote_average'])
    overview.append(i['overview'])
    release_date.append(i['release_date'])
    
d={'title':title,'overview':overview,'release_date':release_date,'popularity':popularity,'vote_count':vote_count,'vote_average':vote_average,'original_language':original_language}
df=pd.DataFrame(d)
    
#  movie title , vote avg
```

```python
final=pd.DataFrame()

for j in range (1,501):        # 500 pages.
    requests.get("https://api.themoviedb.org/3/movie/popular?api_key=3b76c943814293d6a2e91698a5a8bfc1&language=en-US&page={}".format(j)).json()
# appending to old dataframe
    popularity=[]
    vote_count=[]
    title=[]
    original_language=[]
    vote_average=[]
    overview=[]
    release_date=[]
    for i in x['results']:
        popularity.append(i['popularity'])
        vote_count.append(i['vote_count'])
        title.append(i['title'])
        original_language.append(i['original_language'])
        vote_average.append(i['vote_average'])
        overview.append(i['overview'])
        release_date.append(i['release_date'])

    d={'title':title,'overview':overview,'release_date':release_date,'popularity':popularity,'vote_count':vote_count,'vote_average':vote_average,'original_language':original_language}
    df=pd.DataFrame(d)
    final=final.append(df,ignore_index=True)
    
#      hitting on 500 dataframe
```
